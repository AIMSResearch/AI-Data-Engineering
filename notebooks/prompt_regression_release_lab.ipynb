# Prompt Regression Release Lab

This notebook is a cross-chapter prompt and context release lab. It combines evaluation governance from Chapter 6 with release-gate discipline from Chapter 7 by comparing benchmark conversations, prompt versions, corpus versions, retrieval state, tool versions, failure categories, and rollback evidence before a prompt-centered release moves forward.

## Suggested review flow

1. Define benchmark conversations with expected behavior, citation, authorization boundary, and tool-use expectation.
2. Classify each failure mode and contract under pressure.
3. Group regression results by severity and failure category.
4. Compare automated and manual evaluation modes.
5. Produce a release decision and rollback manifest.
6. Reconstruct the evidence bundle for one disputed answer.

## Part 2: Add Failure Taxonomy

A release lab becomes more useful when failures are grouped into a reusable taxonomy instead of being treated as isolated bad answers.

```python
failure_taxonomy = pd.DataFrame([
    {'failure_category': 'hallucinated_policy', 'contract_under_pressure': 'evidence / evaluation', 'description': 'Assistant invents unsupported policy'},
    {'failure_category': 'stale_context', 'contract_under_pressure': 'serving / storage', 'description': 'Assistant uses old retrieved context'},
    {'failure_category': 'unauthorized_context', 'contract_under_pressure': 'governance / serving', 'description': 'Assistant exposes data user should not access'},
    {'failure_category': 'unsafe_tool_use', 'contract_under_pressure': 'governance / evidence', 'description': 'Tool called without valid precondition'},
    {'failure_category': 'unsupported_answer', 'contract_under_pressure': 'evaluation / evidence', 'description': 'Answer is not supported by retrieved chunks'},
])
failure_taxonomy
```

## Part 3: Run Regression Results

The point of a regression suite is not only to list bad cases. It is to turn them into a release decision with traceable reasons.

```python
summary_df = (
    results_df.groupby(['severity', 'contract_under_pressure', 'failure_category', 'pass_fail'])
    .size()
    .reset_index(name='cases')
    .sort_values(['severity', 'failure_category'])
)
critical_failures = int(((results_df['severity'] == 'critical') & (results_df['pass_fail'] == 'fail')).sum())
high_failures = int(((results_df['severity'] == 'high') & (results_df['pass_fail'] == 'fail')).sum())
moderate_failures = int(((results_df['severity'] == 'moderate') & (results_df['pass_fail'] == 'fail')).sum())

if critical_failures > 0:
    release_outcome = 'pause'
elif high_failures > 0:
    release_outcome = 'canary only'
elif moderate_failures > 0:
    release_outcome = 'proceed with warning'
else:
    release_outcome = 'proceed'

print(f'critical_failures = {critical_failures}')
print(f'high_failures = {high_failures}')
print(f'moderate_failures = {moderate_failures}')
print(f'release_outcome = {release_outcome}')
summary_df
```

## Part 4: Automated vs. Manual Evaluation

Release labs usually mix several evaluation modes. Some are easy to automate. Some remain policy or risk judgments that require human review.

```python
evaluation_modes = pd.DataFrame([
    {'evaluation_mode': 'Rule-based checks', 'best_for': 'required citations, tool schema, authorization flags', 'limitation': 'misses nuance'},
    {'evaluation_mode': 'Automated similarity or groundedness', 'best_for': 'scalable regression screening', 'limitation': 'needs calibration'},
    {'evaluation_mode': 'LLM-as-judge', 'best_for': 'qualitative review at scale', 'limitation': 'needs human audit'},
    {'evaluation_mode': 'Human review', 'best_for': 'policy judgment, ambiguity, high-risk cases', 'limitation': 'slower and expensive'},
])
evaluation_modes
```

## Part 5: Human Review Rubric

Human review is most useful when the criteria are explicit enough that two reviewers can explain why they agreed or disagreed.

```python
human_review_rubric = pd.DataFrame([
    {'criterion': 'Groundedness', 'pass': 'Answer supported by retrieved context', 'fail': 'Unsupported claim'},
    {'criterion': 'Policy accuracy', 'pass': 'Uses current policy', 'fail': 'Uses stale or invented policy'},
    {'criterion': 'Authorization', 'pass': 'Protects restricted data', 'fail': 'Exposes unauthorized data'},
    {'criterion': 'Tool use', 'pass': 'Calls tool only when allowed', 'fail': 'Calls tool prematurely'},
    {'criterion': 'Traceability', 'pass': 'Prompt, chunks, and tool output logged', 'fail': 'Missing reconstruction evidence'},
])
human_review_rubric
```

## Part 6: Release Gate

The release gate turns failure severity into a deployment decision. The decision should be explicit enough that the team can explain why the release moved forward, slowed down, or stopped.

```python
release_gate_policy = pd.DataFrame([
    {'outcome': 'proceed', 'condition': 'no high or critical failures'},
    {'outcome': 'proceed with warning', 'condition': 'only moderate failures'},
    {'outcome': 'canary only', 'condition': 'high failure but no critical failure'},
    {'outcome': 'pause', 'condition': 'any critical failure'},
    {'outcome': 'rollback', 'condition': 'critical failure found after exposure'},
])
release_gate_policy
```

## Part 7: Record the Rollback Manifest

If the release is blocked, the rollback target should restore prompt, corpus, retrieval, and tool state together instead of pretending the model alone is the unit of rollback.

```python
rollback_manifest = pd.Series({
    'release_outcome': release_outcome,
    'rollback_prompt_version': 'support_answer_template_v6',
    'rollback_corpus_version': 'support_docs_v13',
    'rollback_retrieval_index_version': 'support_index_v13_4',
    'rollback_tool_version': 'order_status_lookup_v1',
    'benchmark_version': 'support_eval_v9',
    'release_owner': 'support_ai_release_owner',
    'reason': 'critical hallucinated-policy and unauthorized-context failures in the v7 release candidate',
    'approval_timestamp': '2026-04-29T15:00:00Z',
})
rollback_manifest
```

## How to Read the Release Outcome

Use `proceed` only when the suite stays clean enough for the risk level of the application. Use `proceed with warning` when only lower-severity cases fail and the team has accepted the residual risk explicitly. Use `canary only` when higher-severity failures still need controlled exposure. Use `pause` when critical failures remain. The rollback manifest should say exactly which prompt, corpus, retrieval index, and tool versions to restore.
